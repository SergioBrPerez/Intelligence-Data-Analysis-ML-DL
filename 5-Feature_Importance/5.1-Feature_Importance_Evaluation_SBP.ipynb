# 5.1 - Feature Importance Evaluation

## 1: Dependencies and Imports

```python
# Standard library
import os

# Third-party
import joblib
import matplotlib.pyplot as plt
import pandas as pd
```

## 2: Load Models

```python
# Load Clasification Model
model_path = os.path.join('../models', 'lgbm_logistics_model_opt.joblib')
lgbm_model_opt = joblib.load(model_path)

print("Modelo de clasificación cargado correctamente")
```

```python
# Load Regression Model
model_path = os.path.join('../models', 'lgbm_regression_model.joblib')
lgbm_regressor = joblib.load(model_path)

print("Modelo de regresión cargado correctamente")
```

## 3: Data Preparation

```python
# Get Feature Names
feature_names = lgbm_model_opt.feature_name_

# Get Feature Importances
importances = lgbm_model_opt.feature_importances_

# Create sorted DataFrame
fi_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Plot
plt.figure()
plt.barh(fi_df['feature'], fi_df['importance'])
plt.gca().invert_yaxis()
plt.title('Feature Importance - Clasification')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# Get Feature Names
feature_names = lgbm_regressor.feature_name_

# Get Feature Importances
importances = lgbm_regressor.feature_importances_

# Create sorted DataFrame
fi_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Plot
plt.figure()
plt.barh(fi_df['feature'], fi_df['importance'])
plt.gca().invert_yaxis()
plt.title('Feature Importance - Regression')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
